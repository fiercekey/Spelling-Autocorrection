This function is made using the levenshein algorithm, where we can imagine the distances of each edit in a 2d matrix and the [lastrow][last col] will give the minimum edits required.
It is a recursive function that is going to find the minimum of the edits in the cells towards it's left, diagonally left and above it and then return the minimum distance as it's distance

```python
def edit_distance(str1, str2):
    
    def dist(i, j):
        # base cases
        if min(i, j) ==0:
            return max(i, j)
        
        #if the values are same, then the edit distance won't change
        if str1[i - 1] == str2[j - 1]:
            return dist(i - 1, j - 1)
        else:
            return 1 + min(dist(i - 1, j), dist(i, j - 1), dist(i - 1, j - 1))
        
        
    return dist(len(str1), len(str2))
```

Dictionary is the provided list of correctly spelled words

```python
dictionary = np.array(['Doctor', 'Lawyer', 'Teacher', 'Engineer', 'Accountant', 'Nurse', 'Police',  'Architect', 'Dentist', 'Pharmacist'])
dictionary
```

```python
def corrected_str(string, dictionary):
        minimum = float('inf')
        index = 0
        for i in range(len(dictionary)):
            if edit_distance(string, dictionary[i])<minimum:
                minimum = edit_distance(string, dictionary[i])
                index = i
        return dictionary[index]
            
```

```python
def corrected_array(faulty_data, dictionary):
    corrected = []
    for i in range(len(faulty_data)):
        corrected.append(corrected_str(faulty_data[i], dictionary))
    return corrected
```

```python
def frequency(corrected_array, dictionary):
    frequency=[]
    for i in range(len(dictionary)):
        count = 0
        for j in range(len(corrected_array)):
            if corrected_array[j]==dictionary[i]:
                count+=1
        frequency.append(count)
    return frequency
```

```python
def plotting(faulty_data, dictionary):
    corrected_data = corrected_array(faulty_data, dictionary)
    freq = frequency(corrected_data, dictionary)
    plt.bar(dictionary, freq, color ='blue')
    plt.show()
```
